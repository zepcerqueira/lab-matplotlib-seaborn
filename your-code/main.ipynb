```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

average_steps_month = fitbit_db.groupby("Months_encoded")["Steps"].mean()
average_steps_month.name = "Average Steps by Month"
average_steps_month = average_steps_month.reset_index()

average_steps_week = fitbit_db.groupby("Days_encoded")["Steps"].mean()
average_steps_week.name = "Average Steps by Week Days"
average_steps_week = average_steps_week.reset_index()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

fig = make_subplots(rows=1, cols=2, subplot_titles=("Avg. Steps by Month", "Avg. Steps by Week Days"))
fig.add_trace(go.Bar(x=average_steps_month["Average Steps by Month"], name="Average Steps by Month"), row=1, col=1)
fig.add_trace(go.Bar(x=average_steps_week["Average Steps by Week Days"], name="Average Steps by Week Days"), row=1, col=2)

#avg steps is higher on saturday but decreses on sunday. stays arround 10k during weeke days
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
new_new_titanic_db = titanic_db.dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(new_new_titanic_db["Age"], bins=20, edgecolor='k', color='skyblue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("The distribution of column - Age")
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here
uniq_values = new_new_titanic_db["Age"].unique()
```

```python
num_subplots = len(uniq_values)
```

```python

fig, axes = plt.subplots(nrows=num_subplots, ncols=1, figsize=(10, 6))


for i, value in enumerate(uniq_values):
    data_subset = new_new_titanic_db[new_new_titanic_db['Age'] == value]
    ax = axes[i]
    ax.hist(data_subset['Age'], bins=10)
    ax.set_title(value)

fig.suptitle('Histograms of Column_to_Plot Based on Column_Name')
fig.text(0.5, 0.04, 'Column_to_Plot', ha='center')
fig.text(0.04, 0.5, 'Frequency', va='center', rotation='vertical')

    
```
